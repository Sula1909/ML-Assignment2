```python
import pandas as pd
import mlflow
import mlflow.sklearn
import dagshub

dagshub.init(repo_owner='Sula1909', repo_name='ML-Assignment2', mlflow=True)
mlflow.set_tracking_uri('https://dagshub.com/Sula1909/ML-Assignment2.mlflow')

test_transaction = pd.read_csv('/kaggle/input/competitions/ieee-fraud-detection/test_transaction.csv')
test_identity = pd.read_csv('/kaggle/input/competitions/ieee-fraud-detection/test_identity.csv')
test_raw = test_transaction.merge(test_identity, how='left', on='TransactionID')

model_name = 'LGBM_Pipeline'
model_version = '4'

pipeline = mlflow.sklearn.load_model(f"models:/{model_name}/{model_version}")
X_test = test_raw.drop(columns=['TransactionID'], errors='ignore')

predictions = pipeline.predict_proba(X_test)[:, 1]

submission = pd.DataFrame({
    'TransactionID': test_raw['TransactionID'],
    'isFraud': predictions
})
submission.to_csv('submission.csv', index=False)
print('Submission saved!')
submission.head()
```
